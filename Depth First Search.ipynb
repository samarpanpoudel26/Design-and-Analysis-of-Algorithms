```python
class Graph:
    def __init__(self,vertices):
        self.vertices=vertices
        self.adj_list={v: [] for v in range(vertices)} #creating list of vertices
    def add_edge(self,v,w): # adding edges
        self.adj_list[v].append(w)
        self.adj_list[w].append(v)

class DFS:
    def __init__(self,graph,start):
        self.graph=graph
        self.start=start
        self.marked=[False] * graph.vertices
        self.edge_to = [-1]* graph.vertices
        self.dfs(start)
    def dfs(self,v):
        self.marked[v]=True #checking visited or not
        print(f"Visited : {v}")
        for w in self.graph.adj_list[v]: 
            if not self.marked[w]:
                self.edge_to[w]=v
                self.dfs(w)
    def has_path_to(self,v):
        return self.marked[v]
    def path_to(self,v):
        if not self.has_path_to(v):
            return None
        path=[] #creating list of path
        current=v
        while current!=self.start:
            path.append(current)
            current=self.edge_to[current]
        path.append(self.start)
        return path[::-1]

graph= Graph(8)
edges=[(0,1),(0,2),(1,3),(1,4),(2,5),(4,6),(5,7)]

for v,w in edges:
    graph.add_edge(v,w)

dfs = DFS(graph, 0)

print("\n Edge to array:",dfs.edge_to)
print("\n Path from 0 to 6:", dfs.path_to(6))
print("\n Path from 0 to 7:", dfs.path_to(7))        
```
