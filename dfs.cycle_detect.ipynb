## dfs code to detect cycle

```python
class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.adj_list = {v: [] for v in range(vertices)}

    def add_edge(self, v, w):
        self.adj_list[v].append(w)
        self.adj_list[w].append(v)  # Undirected graph


class CycleDetection:
    def __init__(self, graph):
        self.graph = graph
        self.marked = [False] * graph.vertices
        self.has_cycle = False
        
        for v in range(graph.vertices):
          if not self.marked[v]:
            self.dfs(v,-1)
    def dfs(self,v,parent):
      self.marked[v]=True
      for w in self.graph.adj_list[v]:
        if not self.marked[w]:
          self.dfs(w,v)
        elif w!=parent:
          self.has_cycle=True
          
# Example Usage

graph = Graph(8)

edges = [
    (0, 1),
    (0, 2),
    (2, 5),
    (1, 3),
    (1, 4),
    (4, 6),
    (5, 7),
    (6, 1)  # Adding this edge creates a cycle
]

for v, w in edges:
    graph.add_edge(v, w)

cycle_check = CycleDetection(graph)

if cycle_check.has_cycle:
    print("Graph contains a cycle.")
else:
    print("Graph does NOT contain a cycle.")
```
