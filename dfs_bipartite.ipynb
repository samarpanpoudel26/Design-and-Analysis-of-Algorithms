## DFS Function to check if it is bipartite graph or not

```python
class Graph:
  def __init__(self,vertices):
    self.vertices=vertices
    self.adj_list={v:[]for v in range(vertices)}
  def add_edge(self,v,w):
    self.adj_list[v].append(w)
    self.adj_list[w].append(v)

def is_bipartite(graph):
  color=[-1]* graph.vertices # -1 = not colored yet
  def dfs(node):
    for nei in graph.adj_list[node]:
      if color[nei]==-1:
        color[nei]=1-color[node] # give opposite color
        if not dfs(nei):
          return False 
      elif color[nei]==color[node]:
        return False # same color → problem
    return True
  for i in range(graph.vertices):
    if color[i]==-1:
      color[i]=0 # start with any color
      if not dfs(i):
        return False
    return True
    
# Example Usage
graph = Graph(4)
edges = [
    (0, 1),
    (0, 3),
    (1, 2),
    (2, 3)
]

for v, w in edges:
    graph.add_edge(v, w)

if is_bipartite(graph):
    print("The graph is bipartite.")
else:
    print("The graph is NOT bipartite.")
  
```
