## Code to find count dfs connected component

```python
class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.adj_list = {v: [] for v in range(vertices)}

    def add_edge(self, v, w):
        self.adj_list[v].append(w)
        self.adj_list[w].append(v)  # Undirected graph


class ConnectedComponents:
    def __init__(self, graph):
        self.graph = graph
        self.marked = [False] * graph.vertices
        self.id = [-1] * graph.vertices  # Component ID of each vertex
        self.count = 0  # Number of connected components

        for v in range(graph.vertices):
            if not self.marked[v]:
                self.dfs(v)
                self.count += 1
                
    def dfs(self,v):
      self.marked[v]=True
      self.id[v]=self.count
      for w in self.graph.adj_list[v]:
        if not self.marked[w]:
          self.dfs(w)
          
    def connected(self,v,w):
      return self.id[v]==self.id[w]
      

# Example Usage

graph = Graph(8)

edges = [
    (0, 1),
    (0, 2),
    (1, 3),
    (4, 5),
    (5, 6)
    # Vertex 7 is isolated
]

for v, w in edges:
    graph.add_edge(v, w)

cc = ConnectedComponents(graph)

print("Number of connected components:", cc.count)
print("Component IDs:", cc.id)

print("Are 0 and 3 connected?", cc.connected(0, 3))
print("Are 0 and 6 connected?", cc.connected(0, 6))
```
