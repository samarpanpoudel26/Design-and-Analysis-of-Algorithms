## Code for preorder, postorder and reverse order

```python
class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.adj_list = {v: [] for v in range(vertices)}

    def add_edge(self, v, w):
        self.adj_list[v].append(w)  # Directed edge


class DFSOrder:
  def __init__(self,graph):
    self.graph=graph 
    self.marked=[False]*graph.vertices
    self.preorder=[]
    self.postorder=[]
    self.reverse_post=[]
    
    for v in range(graph.vertices):
      if not self.marked[v]:
        self.dfs(v)
        
    self.reverse_post=self.postorder[::-1]
  def dfs(self,v):
    self.marked[v]=True
    
    self.preorder.append(v)
    
    for w in self.graph.adj_list[v]:
      if not self.marked[w]:
        self.dfs(w)
    self.postorder.append(v)
    
    
# Example Usage

graph = Graph(6)

edges = [
    (0, 1),
    (0, 2),
    (1, 3),
    (2, 3),
    (3, 4),
    (4, 5)
]

for v, w in edges:
    graph.add_edge(v, w)

order = DFSOrder(graph)

print("Preorder:", order.preorder)
print("Postorder:", order.postorder)
print("Reverse Postorder:", order.reverse_post)
```
